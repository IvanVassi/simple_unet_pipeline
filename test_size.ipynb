# Тестирование классификатора
Когда мы говорим о тестировании классификатора, мы хотим понять, насколько хорошо он работает. Для этого мы проверяем, насколько его точность (p1) отличается от базовой точности (p0). Но чтобы быть уверенными в результате, нам нужно правильно выбрать количество примеров для теста, то есть размер выборки.
## Основные понятия
1. **Точность классификатора (p1)**: Это насколько хорошо наш классификатор предсказывает правильно. В нашем случае это 0.85, то есть 85%.
2. **Базовая точность (p0)**: Это точность случайного угадывания или простого метода. В нашем случае это 0.5, то есть 50%.
3. **Мощность теста (1 - β)**: Это вероятность, что наш тест правильно покажет, что наш классификатор лучше случайного угадывания. Если мощность 0.95, то это означает, что с вероятностью 95% наш тест найдет разницу, если она действительно существует.
4. **Уровень значимости (α)**: Это вероятность того, что наш тест ошибочно покажет, что наш классификатор лучше, когда на самом деле это не так. Обычно это 0.05, то есть 5%.
## Как это работает?
1. **Средняя точность ($\hat{p}$)**: Мы берем среднее значение между p1 и p0, чтобы понять, как сильно отличается наш классификатор от базового метода. В нашем случае это:
   
   $$\hat{p} = \frac{0.85 + 0.5}{2} = 0.675$$

2. **Квантили нормального распределения**: Нормальное распределение - это график, похожий на колокол, который показывает, как часто встречаются значения. Для уровня значимости 0.05 (то есть 5%) и мощности 0.95 (то есть 95%) мы используем специальные значения (квантили):
   - $Z_{\alpha/2} \approx 1.96$ для уровня значимости.
   - $Z_{\beta} \approx 1.645$ для мощности.

3. **Расчет размера выборки**: Мы используем эти значения и формулу, чтобы узнать, сколько примеров нам нужно, чтобы быть уверенными в результате. Формула выглядит так:

   $$n = \frac{(1.96 \cdot 0.662 + 1.645 \cdot 0.614)^2}{(0.85 - 0.5)^2}$$

4. **Подстановка значений**:
   - $\sqrt{2 \cdot 0.675 \cdot (1 - 0.675)} \approx 0.662$
   - $\sqrt{0.85 \cdot (1 - 0.85) + 0.5 \cdot (1 - 0.5)} \approx 0.614$
   Подставляем все в формулу и считаем:

   $$n = \frac{(1.96 \cdot 0.662 + 1.645 \cdot 0.614)^2}{(0.85 - 0.5)^2} = \frac{(1.29652 + 1.00923)^2}{0.35^2} = \frac{(2.30575)^2}{0.1225} \approx \frac{5.316}{0.1225} \approx 43.4$$

Это означает, что нам нужно примерно 44 примера для теста, чтобы быть уверенными в результате.
## Итог
Чтобы проверить, насколько хорошо работает ваш классификатор, нужно правильно выбрать количество примеров для теста. Мы используем специальные значения и формулы, чтобы рассчитать это количество. В вашем случае, для точности 85% и базовой точности 50%, с мощностью 95%, нужно примерно 44 примера.
